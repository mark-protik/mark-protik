#**Predict crops**

```python
import ipywidgets as widgets
from IPython.display import display
from sklearn.preprocessing import StandardScaler
```

```python
# Create input widgets for each feature
feature_names = x.columns
input_widgets = {feature: widgets.FloatText(description=f"{feature}:") for feature in feature_names}

# Create predict button and output area
predict_button = widgets.Button(description='Predict Crop')
output = widgets.Output()

# Define prediction function
def predict_crop(button):
    with output:
        output.clear_output()
        # Collect input values
        features = [input_widgets[feature].value for feature in feature_names]
        features_array = np.array([features])
        # Scale features if necessary
        if best_model_name in ['Logistic Regression', 'KNN']:
            features_scaled = scaler.transform(features_array)
            prediction_enc = best_model.predict(features_scaled)[0]
        else:
            prediction_enc = best_model.predict(features_array)[0]
        # Decode the prediction
        prediction = le.inverse_transform([prediction_enc])[0]
        print(f"The recommended crop is: {prediction}")

# Attach the prediction function to the button
predict_button.on_click(predict_crop)

# Display the interface
print("\nEnter the values below to predict the recommended crop:")
display(widgets.VBox(list(input_widgets.values()) + [predict_button, output]))
```
