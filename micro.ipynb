#**Micro Project**  

#**Load data**

```python
import pandas as pd
df = pd.read_csv('/content/dataSet.csv')
df
```

##**Data preparation**


###data separation as X and Y

```python
y = df['label']
y
```

```python
x = df.drop('label',axis=1)
x
```

##**Split the dataSet**

###Spliting for test and training

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=100)
```

```python
x_train
```

```python
x_test
```

#**Model Build**

##***Linear Regression***


### **Training the model**

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression


le = LabelEncoder()
y_train_encoded = le.fit_transform(y_train)
y_test_encoded = le.transform(y_test)


lr = LinearRegression()
lr.fit(x_train, y_train_encoded)
```

### **Apply the model to make prediction**

```python
y_lr_train_pred = lr.predict(x_train)
y_lr_test_pred = lr.predict(x_test)
```

```python
y_lr_train_pred
```

```python
y_lr_test_pred
```

### **Evaluate model performance**

```python
y_train_encoded
```

```python

y_pred_int = y_lr_train_pred.round().astype(int)


y_pred_labels = le.inverse_transform(y_pred_int)
y_pred_labels
```

### **Now evaluating**

```python
from sklearn.metrics import mean_squared_error, r2_score

lr_train_mse = mean_squared_error(y_train_encoded, y_lr_train_pred)
lr_train_r2 = r2_score(y_train_encoded, y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test_encoded, y_lr_test_pred)
lr_test_r2 = r2_score(y_test_encoded, y_lr_test_pred)
```

##**Pre processing**

```python
le = LabelEncoder()
y_train_enc = le.fit_transform(y_train)
y_test_enc  = le.transform(y_test)
```

##***Random Forest***

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix


rf = RandomForestClassifier(
    n_estimators=100,       # number of trees
    max_depth=None,         # let trees expand until all leaves are pure
    random_state=100,
    n_jobs=-1               # use all CPU cores
)
rf.fit(x_train, y_train_enc)

y_pred_enc = rf.predict(x_test)

y_pred = le.inverse_transform(y_pred_enc)
```

###**Evaluate the model**

```python
acc = accuracy_score(y_test_enc, y_pred_enc)
print(f"Random Forest Accuracy: {acc:.3f}\n")
```

###**Classificaton Report**

```python
print("Classification Report:")
print(classification_report(y_test_enc, y_pred_enc, target_names=le.classes_))
```

###**Confusion matrix**

```python
print("Confusion Matrix:")
print(confusion_matrix(y_test_enc, y_pred_enc))
```

##***DecisionTreeClassifier***

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix


dt = DecisionTreeClassifier(
    criterion='gini',
    max_depth=None,         # expand until all leaves are pure
    random_state=100
)
dt.fit(x_train, y_train_enc)

y_pred_enc = dt.predict(x_test)

y_pred = le.inverse_transform(y_pred_enc)
```

###**Evaluate the model**

```python
print("Accuracy:", accuracy_score(y_test_enc, y_pred_enc))
```

###**Classification report**

```python
print("\nClassification Report:\n", classification_report(y_test_enc, y_pred_enc))
```

###**Confusion matrix**

```python
print("\nConfusion Matrix:\n", confusion_matrix(y_test_enc, y_pred_enc))
```

##***KNN***

```python

from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

knn = KNeighborsClassifier(
    n_neighbors=5,         # Number of neighbors to use
    metric='minkowski',    # Distance metric (default is Euclidean)
    p=2                    # p=2 for Euclidean, p=1 for Manhattan
)
knn.fit(x_train, y_train_enc)

y_pred_enc = knn.predict(x_test)

y_pred = le.inverse_transform(y_pred_enc)
```

###**Evaluate the model**

```python
print("Accuracy:", accuracy_score(y_test_enc, y_pred_enc))
```
